```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
df=pd.read_csv("/content/newdata.csv")
print(df)
```

```python
#print(df.columns[1],df.columns[2])
data=list(zip(df['Duration'],df['Pulse']))
kmeans = KMeans(n_clusters=5, max_iter=50)
kmeans.fit(data)
```

```python
KMeans(max_iter=50, n_clusters=5)
```

```python

plt.scatter(df['Duration'], df['Pulse'], c=kmeans.labels_)
plt.xlabel('Duration')
plt.ylabel('Pulse')
plt.show()
```
