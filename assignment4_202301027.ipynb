```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/company_sales_data.csv")

plt.plot(df['month_number'],df['total_profit'])

plt.title('Total Profit of All Months')
plt.xlabel('month_number')
plt.ylabel('total_profit')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/company_sales_data.csv")

plt.plot(data['facecream'])
plt.plot(data['facewash'])
plt.plot(data['toothpaste'])
plt.plot(data['bathingsoap'])
plt.plot(data['shampoo'])
plt.title("multiline plot")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/company_sales_data.csv")
plt.bar(data['month_number'],data['total_profit'])
plt.title("bar chart")
plt.xlabel('month_number')
plt.ylabel('total_profit')
plt.show()
```
